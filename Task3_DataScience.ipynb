# State wise Profit Analysis

```python
data_state = data.groupby(['State'])[['Profit','Discount']].mean()
data_state
```

```python
data_state.sort_values("Profit")[["Profit"]].plot(kind='bar', figsize=(10,10))
plt.title("State wise Profit", fontsize=14)
plt.ylabel("Profit Amount")
plt.xlabel("States")
plt.show()
```

# State wise Discount Analysis

```python
data_state.sort_values("Discount")[["Discount"]].plot(kind='bar', figsize=(10,10))
plt.title("State wise Discount", fontsize=14)
plt.ylabel("Discount Amount")
plt.xlabel("States")
plt.show()
```

```python
# Determining Loss States
data_state[data_state['Profit'] < 0]
```

### States are experiencing loss along with the discount provided to such states

```python
data_city = data.groupby(['City'])[['Sales','Discount','Profit']].mean()
data_city = data_city.sort_values(['Profit'])
data_city
```

# Top 15 cities in terms of Profit

```python
data_city['Profit'].tail(15).plot(kind='bar', figsize=(10,10), color = 'Green')
plt.title("Top 15 Cities in terms of Profit", fontsize=15)
plt.ylabel("Total Profit")
plt.xlabel("Cities")
plt.show()
```

# 15 Cities with least Profits

```python
data_city['Profit'].head(15).plot(kind='bar', figsize=(10,10), color = 'Red')
plt.title("Top 15 Cities in terms of Profit", fontsize=15)
plt.ylabel("Total Profit")
plt.xlabel("Cities")
plt.show()
```

# Category Wise Analysis of Sales, Discount and Profit

# Sub Category Wise Analysis of Sales, Discount and Profit

```python
data_sub_category = data.groupby(['Sub-Category'])[['Sales','Discount','Profit']].mean()
data_sub_category
```

```python
plt.figure(figsize=(20,20))
plt.pie(data_sub_category['Sales'], labels = data_sub_category.index, autopct = '%1.1f%%')
plt.title("Sub-Category Wise Sales", fontsize=20)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

### Maximum Sales is in Copiers and Minimum Sales is in Fasteners

# Sub-Category based on Profit

```python
data_sub_category.sort_values('Profit')[['Sales','Profit']].plot(kind='bar', figsize=(10,10))
plt.show()
```

# Sub-Category based on Discount

```python
plt.figure(figsize=(15,15))
plt.pie(data_sub_category['Discount'], labels = data_sub_category.index, autopct = '%1.1f%%')
plt.title("Sub-Category Wise Discount", fontsize=20)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

### Maximum discount is given on Binders and Minimum discount is given on Labels
